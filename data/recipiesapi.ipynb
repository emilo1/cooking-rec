```python
# searchTerm = 'papaya'
# searchURL = f"https://www.amazon.com/s?k={searchTerm}" 
searchTerm = 'papaya'
whole_foods_URL= "https://www.wholefoodsmarket.com"
searchURL = f"{whole_foods_URL}/search?text={searchTerm}"
response = requests.get(searchURL)
alphabet_soup= BeautifulSoup(response.text,"html.parser")
# int(alphabet_soup.find("h1",class_="w-cms--font-headline__serif").text.split(" ")[1])
product_suffix= alphabet_soup.find("div","w-pie--products-grid").find("div", "w-pie--product-tile").find("a")["href"]
product_URL = f"{whole_foods_URL}{product_suffix}"


response = requests.get(product_URL)
alphabet_soup= BeautifulSoup(response.text,"html.parser")
nutrition_facts_html = alphabet_soup.find("section", "w-pie--nutrition-facts")#.find_all("div","nutrition-row")
percent_divs = nutrition_facts_html.find_all("div", "nutrition-column text-bold text-right")
label_divs= nutrition_facts_html.find_all("span", "text-indent")




# label_list = []
# percent_list = []
# for i in range(0,len(percent_divs)):
#     label_list.append(label_divs[i].text)
#     percent_list.append(float(percent_divs[i].text.split("%")[0]))
    
    
# nutrition_facts = {}
# nutrition_facts["label"]=label_list
# nutrition_facts["percent"]=percent_list 
# nutrition_facts

nutrition_list = []
for i in range(0,len(percent_divs)):
    nutrition_list.append({
        "label":label_divs[i].text,
        "percent" : float(percent_divs[i].text.split("%")[0])
        })
    
    
nutrition_list 
```
